##correlation

```python
# 상관관계 행렬 생성
corr = main_data.corr()

# 상관관계가 0.7 이상인 열만 추출
high_corr_cols = set()
for i in range(len(corr.columns)):
    for j in range(i):
        if corr.iloc[i, j] >= 0.8:
            colname = corr.columns[i]
            high_corr_cols.add(colname)
    
```

```python
# 히트맵 그래프 생성
plt.figure(figsize=(10,7))
sns.heatmap(main_data[high_corr_cols].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
def remove_correlated_columns(df, threshold=0.7):
    corr_matrix = df.corr()
    high_corr_var = np.where(np.abs(corr_matrix) > threshold)
    high_corr_var = [(corr_matrix.columns[x], corr_matrix.columns[y]) for x, y in zip(*high_corr_var) if x != y]
    for pair in high_corr_var:
        column1 = pair[0]
        column2 = pair[1]
        if column1 in df.columns:
            df = df.drop(column1, axis=1)
        elif column2 in df.columns:
            df = df.drop(column2, axis=1)
    return df
```

```python
corr_main_data = remove_correlated_columns(main_data, threshold=0.7)
corr_main_data
```

##학습

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from lightgbm.sklearn import LGBMRegressor

```

```python
pipe_model = Pipeline([
               ('pca', PCA(n_components=2)),
               ('model', RandomForestRegressor())
            ])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

hyper_param = {'model__n_estimators': [50, 100, 200],
              'model__max_depth': [2, 5, 10, None],
              'model__min_samples_split': [2, 5, 10],
              'model__min_samples_leaf': [1, 2, 4],
              'model__max_features': ['auto', 'sqrt', 'log2', None],
              'model__bootstrap': [True, False]}

grid_model = GridSearchCV(pipe_model, param_grid=hyper_param, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
grid_model.fit(X_train, y_train)
best_model = grid_model.best_estimator_
```

```python
y_train_pred = best_model.predict(X_train)
print(mean_squared_error(y_train, y_train_pred))
```

```python
y_test_pred = best_model.predict(X_test)
print(mean_squared_error(y_test, y_test_pred))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_test_pred)
```

```python
y_test_pred_df = pd.DataFrame(y_test_pred, columns=["pred"])
y_test.reset_index(drop=True, inplace=True)
result = pd.concat([y_test, y_test_pred_df], axis=1)
result
```
