```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_path = "/content/drive/MyDrive/Colab Notebooks/data/oes/"
data = pd.read_csv(base_path + '2023-02-01_캡스톤디자인_oes.csv')
data
```

```python
peak_point = data.loc[data['Time(sec)'] == '17h45m03s',:]
```

```python
maximum_point = pd.DataFrame(peak_point.max(axis=0)).T
```

```python
maximum_point = pd.DataFrame(maximum_point.iloc[0,1:])
```

```python
maximum_point
```

```python
plt.figure(figsize=(16,8))
plt.plot(maximum_point.index, maximum_point[0])
total_len = len(maximum_point.index)
plt.xticks(np.arange(0, total_len, 200), rotation=90)
plt.show
```

```python
fig = plt.figure(figsize=(16,10))

ax1 = fig.add_subplot(2,3,1)
sns.lineplot(x=data.index, y=data['357.668'], ax=ax1)
ax1.set_title('N(357)')

ax2 = fig.add_subplot(2,3,2)
sns.lineplot(x=data.index, y=data['380.436'], ax=ax2)
ax2.set_title('N(380)')

ax3 = fig.add_subplot(2,3,3)
sns.lineplot(x=data.index, y=data['391.665'], ax=ax3)
ax3.set_title('N(391)')

ax4 = fig.add_subplot(2,3,4)
sns.lineplot(x=data.index, y=data['777.365'], ax=ax4)
ax4.set_title('O(777)')

ax5 = fig.add_subplot(2,3,5)
sns.lineplot(x=data.index, y=data['749.401'], ax=ax5)
ax5.set_title('Ar(749)')

plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.xlim(8000, 8600)
plt.ylim(4300, 4800)
sns.scatterplot(x=data['357.668'], y=data['380.436'])

plt.subplot(2,2,2)
plt.xlim(8000, 8600)
plt.ylim(4300, 4900)
sns.scatterplot(x=data['357.668'], y=data['777.365'])

plt.subplot(2,2,3)
plt.xlim(2450, 2800)
plt.ylim(7900, 8600)
sns.scatterplot(y=data['357.668'], x=data['749.401'])

plt.show()
```
